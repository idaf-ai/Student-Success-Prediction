```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
file = 'student-mat.csv'
df = pd.read_csv(file, delimiter=';')
pd.set_option('display.max_columns', None)
df.head()
```

```python
df.describe()
```

```python
df_copy = df.copy()
```

```python
def pass_fail(data):
    grades = []
    for scores in data['G3']:
        if scores >= 10:
            scores = 'pass'
        else:
            scores = 'fail'
        grades.append(scores)
    data['finalExamGrade'] = grades
    return data
```

```python
df_new = pass_fail(df_copy)
df_new.head()
```

```python
x = df_new.corr()
```

```python
plt.figure(figsize = (12,6))
sns.heatmap(x, annot=True)
```

### Number of Male vs Female Students who failed in the final exam

```python
attributes = ['sex', 'G3', 'finalExamGrade']
df_fail = df_new.loc[df_new['finalExamGrade']=='fail', attributes]
df_fail.head()
```

```python
sns.barplot(x='sex', y='G3', data=df_fail)
```

```python
sns.barplot(x='sex', y='G3', hue =  'finalExamGrade' ,data=df_new)
#https://files.eric.ed.gov/fulltext/EJ1083613.pdf
# male student tend to do slightly better than female students
```

### Number of Male vs Female Students who passed in the final exam

```python
attributes = ['sex', 'G3', 'finalExamGrade']
df_pass = df_new.loc[df_new['finalExamGrade']=='pass', attributes]
df_pass.head()
```

```python
plt.figure(figsize = (8,10))

plt.subplot(2,1,1)

g = sns.countplot(x='sex', data=df_new)
g.set_title("Number of student", fontsize=15)
g.set_xlabel("sex", fontsize=12)


total = float(len(df_new))
for p in g.patches:
    height = p.get_height()
    g.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.2f}'.format(height/total),
            ha="center") 
    
    

plt.subplot(2,1,2)  
gg = sns.barplot(x='sex', y='G3', data=df_pass)
gg.set_title("Performance by Gender", fontsize=15)
gg.set_xlabel("sex", fontsize=12)
gg.set_ylabel("Performance", fontsize=12)

#We have more females than males but the males tend to do slightly better than the females


#plt.savefig('CountofStudent.png')
plt.show
```
